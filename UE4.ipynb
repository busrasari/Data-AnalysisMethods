```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head(10)
```

##(1) Datenverständnis

```python
iris_df.describe()
```

```python
iris_df.shape
```

```python
iris_df.isnull().any()
```

##(2) Korrelationsanalyse 

```python
sns.set_theme()
sns.heatmap(iris_df.corr(), annot = True)
```

##(3) Analyse Ausreißer

```python
iris_df['target'] = iris.target
```

```python
f, axs = plt.subplots(1, 4)
sns.boxplot(x="target", y="sepal length (cm)", data=iris_df, ax = axs[0])
sns.boxplot(x="target", y="petal length (cm)", data=iris_df, ax = axs[1])
sns.boxplot(x="target", y="petal width (cm)", data=iris_df, ax = axs[2])
sns.boxplot(x="target", y="sepal width (cm)", data=iris_df, ax = axs[3])
plt.tight_layout()
plt.show()
```

##(4) Explorative Datenanalyse (EDA)

```python
sns.pairplot(iris_df[['sepal length (cm)', 'petal length (cm)', 'petal width (cm)','sepal width (cm)']])
```

```python
sns.scatterplot(iris_df['sepal length (cm)'], iris_df['petal length (cm)'],hue =iris_df['target'],s=30)
```

```python
b=sns.jointplot(x='sepal length (cm)', y='petal length (cm)', data=iris_df, size=8, alpha=.6,color='k', marker='x')
```
